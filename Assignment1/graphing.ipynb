```python
import matplotlib.pyplot as plt
import os
import re
```

```python
def extract_data(file):
    #Input is the file name which is exp1/subexp1/pur
    #iterate through the files pur8.log, pur32.log, pur128.log, pur512.log, pur2048.log
    #extract the time taken to read the files and time taken by compute part of the function
    #return the two arrays
    read_time = []
    compute_time = []
    for i in [8, 32, 128, 512, 2048]:
        with open(f'{file}{i}.log', 'r') as f:
            lines = f.readlines()
            read_time.append(int(re.search(r'\d+', lines[0]).group()))
            compute_time.append(int(re.search(r'\d+', lines[1]).group()))
    return read_time, compute_time
```

```python
def graph_data(read_time, compute_time, file, exp_no, subexp_no):
    x = [8, 32, 128, 512, 2048]
    plt.plot(x, read_time, label='Time taken to read files')
    plt.plot(x, compute_time, label='Time taken by compute part of function')
    #Add exp_no and subexp_no to the title
    plt.title(f'Experiment {exp_no} Subexperiment {subexp_no}')
    plt.xlabel('Purchase size')
    plt.ylabel('Time taken in microseconds')
    plt.legend()
    plt.savefig(file)
    plt.close()
```

```python
def exp1():
    #The files are stored in exp1\subexp1, exp1\subexp2, exp1\subexp3, and exp1\subexp4
    for i in range(1, 5):
        read_time, compute_time = extract_data(f'exp1/subexp{i}/pur')    
        graph_data(read_time, compute_time, f'exp1graphsubexp{i}.png', 1, i)
```

```python
#Write a new extract data function for the second experiment:
#The data is stored in log files such as exp2/subexp1/customer4pur8.log, exp2/subexp1/customer4pur32.log, exp2/subexp1/customer4pur128.log, exp2/subexp1/customer4pur512.log, exp2/subexp1/customer4pur2048.log, the general form is exp2/subexp1/customer{customer}pur{purchase}.log
#customer is 4, 8, 16, 32, 64,  and purchase is 8, 32, 128, 512, 2048
#extract the time taken to read the files and time taken by compute part of the function
#return the two arrays
#files only exist if customer is less than or equal to purchase
def extract_data2_customer_const(file, customer):
    read_time = []
    compute_time = []
    for i in [8, 32, 128, 512, 2048]:
        if i >= customer:
            with open(f'{file}{customer}pur{i}.log', 'r') as f:
                lines = f.readlines()
                read_time.append(int(re.search(r'\d+', lines[0]).group()))
                compute_time.append(int(re.search(r'\d+', lines[1]).group()))
    return read_time, compute_time

def extract_data2_purchase_const(file, purchase):
    read_time = []
    compute_time = []
    for i in [4, 8, 16, 32, 64]:
        if i <= purchase:
            with open(f'{file}{i}pur{purchase}.log', 'r') as f:
                lines = f.readlines()
                read_time.append(int(re.search(r'\d+', lines[0]).group()))
                compute_time.append(int(re.search(r'\d+', lines[1]).group()))
    return read_time, compute_time
```

```python
#Write a new graph data function for the second experiment:
#The data is stored in log files such as exp2/subexp1/customer4pur8.log, exp2/subexp1/customer4pur32.log, exp2/subexp1/customer4pur128.log, exp2/subexp1/customer4pur512.log, exp2/subexp1/customer4pur2048.log, the general form is exp2/subexp1/customer{customer}pur{purchase}.log
#customer is 4, 8, 16, 32, 64,  and purchase is 8, 32, 128, 512, 2048
#extract the time taken to read the files and time taken by compute part of the function
#return the two arrays
#files only exist if customer is less than or equal to purchase
def graph_data2_customer_const(read_time, compute_time, file, customer, exp_no, subexp_no):
    #The x values should be greater than or equal to customer
    plt.figure(figsize=(10, 5))
    x = [8, 32, 128, 512, 2048]
    x = [i for i in x if i >= customer]
    plt.plot(x, read_time, label='Time taken to read files')
    plt.plot(x, compute_time, label='Time taken by compute part of function')
    #Add the exp number and subexp number to title
    plt.title(f'Time taken to read files and compute part of function vs file size with customer={customer} in exp{exp_no} subexp{subexp_no}')
    plt.xlabel('Purchase size')
    plt.ylabel('Time taken in microseconds')
    plt.legend()
    plt.savefig(file)
    plt.close()

def graph_data2_purchase_const(read_time, compute_time, file, purchase, exp_no, subexp_no):
    #Resize plot to fit title and labels
    plt.figure(figsize=(10, 5))
    #The x values should be less than or equal to purchase
    x = [4, 8, 16, 32, 64]
    x = [i for i in x if i <= purchase]
    plt.plot(x, read_time, label='Time taken to read files')
    plt.plot(x, compute_time, label='Time taken by compute part of function')
    #Add the exp number and subexp number to title
    plt.title(f'Time taken to read files and compute part of function vs file size with purchase={purchase} in exp{exp_no} subexp{subexp_no}')
    plt.xlabel('Customer size')
    plt.ylabel('Time taken in microseconds')
    plt.legend()
    plt.savefig(file)
    plt.close()

def exp2():
    #The files are stored in exp2\subexp1, exp2\subexp2, exp2\subexp3, and exp2\subexp4
    for i in range(1, 5):
        for customer in [4, 8, 16, 32, 64]:
            read_time, compute_time = extract_data2_customer_const(f'exp2/subexp{i}/customer', customer)
            graph_data2_customer_const(read_time, compute_time, f'exp2graphsubexp{i}customer{customer}.png', customer, 2, i)
        for purchase in [128, 512, 2048]:
            read_time, compute_time = extract_data2_purchase_const(f'exp2/subexp{i}/customer', purchase)
            graph_data2_purchase_const(read_time, compute_time, f'exp2graphsubexp{i}purchase{purchase}.png', purchase, 2, i)
```
